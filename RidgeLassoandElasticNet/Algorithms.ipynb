```python
df = pd.read_csv("Algerian_forest_fires_dataset_cleaned.csv")
df.head()
```

```python
df = df.drop(columns=["day","month","year"],axis=1)
df.head()
```

```python
df["Classes"].value_counts()
```

```python
## Encoding
df['Classes']=np.where(df['Classes'].str.contains("not fire"),0,1)
```

```python
df["Classes"].value_counts()
```

```python
## INdependent and dependent features
x= df.drop("FWI",axis=1)
y = df["FWI"]
```

```python
## Train test split 
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,random_state=42,test_size=0.25)
```

```python
##Feature selection based on correlation
```

```python
df.corr()
```

```python
## Check for multicollinearity
plt.figure(figsize=(12,10))
corr = xtrain.corr()
sns.heatmap(corr,annot=True)
```

```python
def correlation(dataset, threshold):
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: 
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
## threshold--Domain expertise
corr_features=correlation(xtrain,0.85)
corr_features
```

```python
## drop features when correlation is more than 0.85 
xtrain.drop(corr_features,axis=1,inplace=True)
xtest.drop(corr_features,axis=1,inplace=True)
xtrain.shape,xtest.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
xtrain_scaled=scaler.fit_transform(xtrain)
xtest_scaled=scaler.transform(xtest)
```

```python
xtrain_scaled
```

```python
plt.subplots(figsize=(15, 5))
plt.subplot(1, 2, 1)
sns.boxplot(data=xtrain)
plt.title('X_train Before Scaling')
plt.subplot(1, 2, 2)
sns.boxplot(data=xtrain_scaled)
plt.title('X_train After Scaling')
```

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
lasso=Lasso()
lasso.fit(xtrain_scaled,ytrain)
y_pred=lasso.predict(xtest_scaled)
mae=mean_absolute_error(ytest,y_pred)
score=r2_score(ytest,y_pred)
print("Mean absolute error", mae)
print("R2 Score", score)
plt.scatter(ytest,y_pred)
```

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
ridge=Ridge()
ridge.fit(xtrain_scaled,ytrain)
y_pred=ridge.predict(xtest_scaled)
mae=mean_absolute_error(ytest,y_pred)
score=r2_score(ytest,y_pred)
print("Mean absolute error", mae)
print("R2 Score", score)
plt.scatter(ytest,y_pred)
```

```python
from sklearn.linear_model import ElasticNet
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
lasso=ElasticNet()
lasso.fit(xtrain_scaled,ytrain)
y_pred=lasso.predict(xtest_scaled)
mae=mean_absolute_error(ytest,y_pred)
score=r2_score(ytest,y_pred)
print("Mean absolute error", mae)
print("R2 Score", score)
plt.scatter(ytest,y_pred)
```
